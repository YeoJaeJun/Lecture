````python
import seaborn as sns
````

````python
import matplotlib.pyplot as plt
````

## 1.1. rugplot, kdeplot, displot
- 1차원 연속형 값들의 분포를 시각화 하는 그래프

### 1.1.1. rugplot 
- 각 데이터들의 위치를 보여준다.

````python
plt.figure(figsize = (4, 3))

sns.rugplot(tips['total_bill'], alpha = 0.5)
plt.title('total_bill의 분포')

plt.show()
````

````python
plt.figure(figsize = (4, 3))

sns.rugplot(x = 'total_bill', data = tips, alpha = 0.5)
plt.title('total_bill의 분포')

plt.show()
````

### 1.1.2. kdeplot
- 히스토그램을 부드러운 곡선 형태로 표현한다. 
- KDE(Kernel Density Estimation) : 확률밀도추정

````python
plt.figure(figsize = (4, 3))

sns.kdeplot(tips['total_bill'])
plt.title('total_bill의 분포')

plt.show()
````

````python
plt.figure(figsize = (4, 3))

sns.kdeplot(x = 'total_bill', data = tips)
plt.title('total_bill의 분포')

plt.show()
````

### 1.1.3. displot()
- 히스토그램을 기준으로 rugplot, kdeplot을 같이 그릴 수 있다.

````python
sns.displot(x = 'total_bill', data = tips, bins = 30)
plt.title('총 금액의 분포')

plt.show()
````

````python
sns.displot(x = 'total_bill', data = tips, bins = 30, 
            rug = True)
plt.title('총 금액의 분포')

plt.show()
````

````python
sns.displot(x = 'total_bill', data = tips, bins = 30, 
            rug = True, kde = True)
plt.title('총 금액의 분포')

plt.show()
````

````python
# 흡연 여부에 따른 총 금액의 분포를 확인해보자.
# 이럴 때는 parameter hue를 이용한다.

sns.displot(x = 'total_bill', hue = 'smoker', data = tips)
plt.title('흡연 여부에 따른 총 금액의 분포')

plt.show()
````

## 1.3. boxplot(), violinplot(), swamplot()
- 연속형 데이터(양적데이터)들의 분포를 확인하는 그래프를 그린다.
- 범주별로 연속형 데이터의 분포를 비교할 수 있다.

### 1.3.1. boxplot

````python
# 총 금액과 팁의 분포를 서로 다른 subplot에 그려보자.

plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x = 'total_bill', data = tips)
plt.title('총 금액의 분포')

plt.subplot(1, 2, 2)
sns.boxplot(y = 'tip', data = tips)
plt.title('팁의 분포')

plt.show()
````

````python
fig = plt.figure(figsize = (10, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.boxplot(x = 'total_bill', data = tips, ax = ax1)
sns.boxplot(y = 'tip', data = tips, ax = ax2)

ax1.set_title('총 금액의 분포')
ax2.set_title('팁의 분포')

plt.show()
````

````python
# 총 금액과 팁의 분포를 성별에 따라 나눠서 그래프로 그려보자.
````

````python
fig = plt.figure(figsize = (10, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.boxplot(x = 'total_bill', y = 'sex', data = tips, ax = ax1)
sns.boxplot(x = 'sex', y = 'tip', data = tips, ax = ax2)

ax1.set_title('성별에 따른 총 금액의 분포')
ax2.set_title('성별에 따른 팁의 분포')

plt.show()
````

````python
# 총 금액과 팁의 분포를 성별과 흡연 여부에 따라 나눠서 그래프로 그려보자. 이때 parameter hue를 이용한다.
````

````python
fig = plt.figure(figsize = (10, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.boxplot(x = 'total_bill', y = 'sex', hue = 'smoker', data = tips, ax = ax1)
sns.boxplot(x = 'sex', y = 'tip', hue = 'time', data = tips, ax = ax2)

ax1.set_title('성별과 흡연 여부에 따른 총 금액의 분포')
ax2.set_title('성별과 시간대에 따른 팁의 분포')

plt.show()
````

````python
# boxplot의 한계점은 분위수의 값의 개수를 확인할 수 없다는 것이다. 이 점을 보완해주는 것이 vilolin plot이다. 
````

### 1.3.2. violin plot
- boxplot 위에 kdeplot를 좌우 대칭으로 덮어씌워 data의 분포를 표현한다. 그래서 boxplot 보다 좀더 정확한 data의 분포를 볼 수 있다.
- parameter는 boxplot과 동일하다.

````python
# 팁의 분포를 violin plot으로 그려보자.

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.violinplot(data = tips, x = 'tip')
plt.title('팁의 분포')

plt.subplot(1, 2, 2)
sns.violinplot(data = tips, y = 'tip')
plt.title('팁의 분포')

plt.show()
````

````python
# 요일별 팁의 분포를 그려보자.

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.violinplot(data = tips, x = 'tip')
plt.title('팁의 분포')

plt.subplot(1, 2, 2)
sns.violinplot(data = tips, x = 'day', y = 'tip')
plt.title('요일별 팁의 분포')

plt.show()
````

````python
# 여러 분류별 팁의 분포를 그려보자.

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.violinplot(data = tips, x = 'tip', y = 'sex', hue = 'smoker')
plt.title('성별과 흡연 여부에 따른 팁의 분포')

plt.subplot(1, 2, 2)
sns.violinplot(data = tips, x = 'day', y = 'tip', hue = 'smoker')
plt.title('요일과 흡연 여부에 따른 팁의 분포')

plt.show()
````

### 1.3.3. swarmplot
- 실제 값들을 점으로 찍어 준다. 
- boxplot이나 violinplot의 보안해주는 역할로 쓰인다.
- swarmplot은 가운데 분류를 기준으로 분포시키는데 실제 값이 있는 위치에 점을 찍어 값이 어디에 있는지 좀 더 정확하게 알 수 있다.

````python
# 팁의 분포를 그려보자.

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.swarmplot(x = 'tip', data = tips)
plt.title('팁의 분포')

plt.subplot(1, 2, 2)
sns.swarmplot(y = 'tip', data = tips)
plt.title('팁의 분포')

plt.show()
````

````python
# 보통 swarmplot은 다른 그래프 위에 덮어쓰는 용도로 사용한다.

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.boxplot(x = 'tip', data = tips)
sns.swarmplot(x = 'tip', data = tips, color = 'orange')
plt.title('팁의 분포')

plt.subplot(1, 2, 2)
sns.boxplot(y = 'tip', data = tips)
sns.swarmplot(y = 'tip', data = tips, color = 'orange')
plt.title('팁의 분포')

plt.show()
````

### 1.3.4. countplot() 
- 막대그래프(bar plot)을 그리는 함수
- 범주형 변수의 고유값의 개수를 표시
- matplotlib의 bar()

````python
# 요일별 tip을 받은 횟수를 그려보자.

plt.figure(figsize = (5, 3))

sns.countplot(data = tips, x = 'day')
plt.title('요일별 팁을 받은 횟수')

plt.show()
````

````python
# 위 vertical bar plot을 horizontal bar plot으로 그려보자.

plt.figure(figsize = (5, 3))

sns.countplot(data = tips, y = 'day')
plt.title('요일별 팁을 받은 횟수')

plt.show()
````

````python
# 요일별 흡연자/비흡연자의 수를 그려보자.

plt.figure(figsize = (5, 3))

sns.countplot(data = tips, x = 'smoker', hue = 'day')
plt.title('요일별 흡연자/비흡연자의 수')
plt.legend(bbox_to_anchor = (1, 1), loc = 'upper left')

plt.show()
````

````python
# 이때 parameter dodge의 기능을 확인해보자.
# dodge = Fasle로 두면 vertical stacked bar plot을 그린다.
# dodge의 default는 True이다. dodge = True로 입력하면 horizontal stacked bar plot을 그린다.

plt.figure(figsize = (5, 3))

sns.countplot(data = tips, x = 'smoker', hue = 'day', dodge = False, alpha = 0.5)
plt.title('요일별 흡연자/비흡연자의 수')
plt.legend(bbox_to_anchor = (1, 1), loc = 'upper left')

plt.show()
````

## 1.4. scatterplot, lmplot, jointplot, pairplot
- 산점도를 그린다.

### 1.4.1. scatterplot

팔레트 - https://seaborn.pydata.org/tutorial/color_palettes.html#palette-tutorial

````python
# scatterplot은 x축에 들어갈 값과 y축에 들어갈 값이 필수로 한다.
# 총 금액과 팁의 관계를 확인해보자.

plt.figure(figsize = (4, 4))

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', alpha = 0.5)
plt.title('총 금액과 팁의 관계')

plt.show()
````

````python
# 흡연여부에 따른 총 금액과 팁의 관계를 그려보자.

plt.figure(figsize = (4, 4))

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', alpha = 0.5)
plt.title('흡연여부에 따른 총 금액과 팁의 관계')

plt.show()
````

````python
plt.figure(figsize = (4, 4))

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', alpha = 0.5, palette = 'Set1')
plt.title('흡연여부에 따른 총 금액과 팁의 관계')

plt.show()
````

### 1.4.2. lmplot()
- 선형회귀 적합선을 포함한 산점도를 그린다.

````python
# 총 금액과 팁의 관계를 그려보자.

sns.lmplot(data = tips, x = 'total_bill', y = 'tip')
plt.title('총 금액과 팁의 관계')

plt.show()
````

````python
# 위 그래프에서 직선이 선형회귀 적합선이고 이 직선을 이용해서 주어진 data를 통해 값을 예측할 수 있다.
````

````python
# 성별에 따른 총 금액과 팁의 관계를 확인해보자.

sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', palette = 'Set1')
plt.title('성별에 따른 총 금액과 팁의 관계')

plt.show()
````

### 1.4.3. jointplot()
- scatter plot 과 각 변수의 히스토그램을 같이 그린다.
- pandas **DataFrame**만 사용할 수 있다.
- DataFrame

````python
# 총 금액과 팁의 관계를 그려보자.

sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
plt.title('총 금액과 팁의 관계', loc = 'left', pad = 80)

plt.show()
````

````python


sns.pairplot(tips)

plt.show()
````

## 1.5. heatmap()
- 값들에 비례해서 색깔을 다르게 해 2차원 자료로 시각화

````python
tips[['total_bill', 'tip', 'size']].corr()
````

````python
# 총 금액, 팁, 손님의 총 수의 상관관계를 heatmap으로 그려보자.

sns.heatmap(tips[['total_bill', 'tip', 'size']].corr(), 
            annot = True, 
            fmt = '.3f', 
            cmap = 'Blues')
plt.title('총 금액, 팁, 손님의 총 수의 상관관계')

plt.show()
````

````python
# annot는 값이 heatmap에 나오도록 하는 것이다. fmt는 출력되는 문자열의 형식을 지정하는 것이고,
# cmap은 palette처럼 style을 지정하는 것이다.
# 색이 진할수록 상관관계가 높은 것이고 연할수록 상관관계가 낮은 것이다.
````

## 1.6. lineplot
- 선그래프
- 시간의 흐름에 따른 값의 변화를 보여주는데 유용하다. (시계열 데이터)

````python
# 먼저 Time Series Data를 생성한다.

import pandas as pd
import numpy as np

index = pd.date_range('2023/01/01', freq = 'MS', periods = 10)
values = np.random.randint(1, 100, size = (10, 3))
df = pd.DataFrame(values, index = index, columns = ['no1', 'no2', 'no3'])
df
````

````python
plt.figure(figsize = (10, 3))

sns.lineplot(data = df, x = df.index, y = 'no1')
plt.title('날짜별 no1 값의 변화')
plt.xlabel('date')
plt.ylabel('value')

plt.show()
````

````python
plt.figure(figsize = (10, 3))

sns.lineplot(data = df, x = df.index, y = 'no1', label = 'no1')
sns.lineplot(data = df, x = df.index, y = 'no2', label = 'no2')
sns.lineplot(data = df, x = df.index, y = 'no3', label = 'no3')
plt.title('날짜별 no1, no2, no3 값의 변화')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()

plt.show()
````

````python
# 기존 df에 문자를 값으로 가지는 cate column을 추가한다.
````

````python
df['cate'] = ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'C']
df
````

````python
# 날짜별 no1 값의 변화를 cate에 따라 분류해보자.

plt.figure(figsize = (10, 3))

sns.lineplot(data = df, x = df.index, y = 'no1', hue = 'cate', marker = 'o')
plt.title('날짜별 no1 값의 변화를 cate에 따라 분류')
plt.xlabel('date')
plt.ylabel('value')

plt.show()
````
