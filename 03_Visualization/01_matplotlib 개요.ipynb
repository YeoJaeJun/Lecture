## 1.2. matplotlib 그래프 구성요소


- **figure**
    - 전체 그래프가 위치할 기본 틀
    - 하나의 figure에 여러개의 그래프를 그릴 수 있다.
    - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
- **axes(subplot)**
    - figure내에 한개의 그래프를 그리기 위한 공간
        - figure에 한개 이상의 axes(subplot)로 구성해서 각 axes에 그래프를 그린다.
    - https://matplotlib.org/stable/api/axes_api.html
- **axis** 
    - 축 (x축, y축)은 값들을 위치시키는 선을 말한다.
    - axis label (x, y) : 축의 레이블(설명)
- **ticks** : 축의 값을 알려주는 눈금
    - Major tick
    - Minor tick
- **title** : 플롯 제목   
- **legend (범례)**
    - 하나의 axes내에 여러 그래프를 그린 경우 그것에 대한 설명

```python
%matplotlib inline
# %matplotlib qt

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(15,7), facecolor='gray') #facecolor: figure의 배경 색
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.plot([1,2,3,4,5], [10,20,30,40,50], label='line1')
axes1.plot([1,2,3,4,5], [50,40,30,20,10], label='line2')
axes2.scatter(np.random.randint(100, size=50), np.random.randint(100, 200, size=50), color='r')

fig.suptitle('Example of Plot', size=25, color='blue') #size: 폰트 크기, color: 폰트 색
axes1.set_title("PLOT 1", size=20)
axes2.set_title("Plot 2", size=20)

axes1.set_xlabel("X축", size=15)
axes1.set_ylabel("Y축", size=15)
axes2.set_xlabel("가격1", size=15)
axes2.set_ylabel('가격2', size=15)

axes1.legend()
axes1.grid(True)
plt.show()
```

## 3.1. pyplot 모듈을 이용해 그리기
1. pyplot 모듈이 그래프 그리는 함수와 Axes(Subplot) 설정 관련 함수를 제공

```python
# 먼저 그래프를 그릴 data를 생성한다. 그래프를 그릴 때는 1차원 형태의 data structure는 모두 사용 가능하다.

x = [1, 2, 3, 4, 5]
y = range(10, 60, 10)
```

```python
# matplotlib.pyplot을 import한다. 관례적으로 plt라는 별칭을 사용한다.

import matplotlib.pyplot as plt

# 1. figure의 크기를 가장 먼저 설정한다. 가로와 세로의 길이를 inch로 설정한다.
plt.figure(figsize = (10, 7))

# 2. 그래프를 그린다.
plt.plot(x, y)

# 3. 추가적인 설정을 한다.
# subplot의 제목 설정
plt.title("Line Graph Example", fontsize = 20)
# 축의 label 설정
plt.xlabel("x-axis values")
plt.ylabel("y-axis values")
# 눈금선 설정
plt.grid(True)

# 참고: 그래프를 저장하는 방법
plt.savefig("sample.png")

# 4. 화면에 출력
plt.show()
```

2. 하나의 figure에 여러 그래프 그리기
    - plt.subplot(row,columns,num)

```python
plt.figure(figsize = (15, 10))

# 그래프를 그릴 subplot을 설정한다. 이는 내가 직접 설정할 수 있다.
# figure를 2 x 3으로 나누고 번호를 메기면 왼쪽 상단 1번 번호가 메겨지고 우측 하단으로 가면서 1씩 증가한다.
# 직접 확인해보자.
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title("Graph 1")

plt.subplot(2, 3, 2)
plt.scatter(x, y)
plt.title("Graph 2")

plt.subplot(2, 3, 3)
plt.bar(x, y)
plt.title("Graph 3")

plt.subplot(2, 3, 4)
plt.pie(x)
plt.title("Graph 4")

plt.subplot(2, 3, 5)
plt.pie(y)
plt.title("Graph 5")

plt.subplot(2, 3, 6)
plt.boxplot(x)
plt.title("Graph 6")

plt.tight_layout()

plt.show()
```

```python
# 한 개의 subplot에 여러 그래프 그리는 방법
plt.plot(x, y, label = "graph 1")
plt.scatter(x, y, label = "graph 2")
plt.plot(y, x, label = "graph 3")

plt.legend()

plt.show()
```

## 3.2. Figure 와 Axes 객체를 이용해 그리기

- AxesSubplot 객체(Axes를 표현하는 객체)의 메소드들을 이용해 그래프를 그린다.
- axes 생성 방법
    - plt.gca(): AxesSubplot
        - 하나의 figure에 하나의 Axes 만 사용할 경우 사용.
    - 하나의 figure에 여러개의 Axes들을 사용할 경우
        - figure.add_subplot() 메소드 이용
            - figure를 먼저 생성후 axes 들을 추가
        - pyplot.subplots() 함수를 이용
            - figure와 axes배열을 동시에 생성

### 3.2.1. Figure에 하나의 axes 그리기
- plt.gca() 사용

```python
# plt.gca() 사용 방법
# axes instance의 method를 이용해 그래프를 그리고 설정한다.

axes = plt.gca()
axes.plot(x, y, label = "graph 1")
axes.scatter(x, y, label = "graph 2")

axes.set_title("Graph")
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
axes.legend()
axes.grid(True)

plt.show()
```

### 3.2.2. 하나의 Figure에 여러개의 axes 그리기
1. <b style='font-size:1.3em'>figure.add_subplot() 메소드 이용</b>
    - figure객체에 axes를 추가하는 형태
    - nrows(총행수), ncols(총열수), index(axes위치) 지정

```python
# 먼저 figure instance를 조회해야 한다.

fig = plt.figure(figsize = (10, 5))

# 그 후 figure에 axes를 추가한다.
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# 각 axes의 그래프를 그린다.
ax1.plot(x, y)
ax2.scatter(x, y)
ax3.plot(x, y)

ax1.grid(True)
ax3.grid(True)

plt.tight_layout()

plt.show()
```

2. <b style='font-size:1.3em'>pyplot.subplots()</b>
    - nrows, ncols 로 axes개수와 위치 지정
    - 반환: figure와 axes(subplot)들을 담은 **ndarray**

```python
fig, axes = plt.subplots(1, 2, figsize = (7, 3))

axes[0].plot(x, y)
axes[0].set_title("Graph 1")

axes[1].plot(x, y)
axes[1].set_title("Graph 2")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (7, 3))

ax1, ax2 = axes
ax1.plot(x, y)
ax1.set_title("Graph 1")

ax2.plot(x, y)
ax2.set_title("Graph 2")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (10, 10))

ax1, ax2, ax3, ax4 = axes.flatten()

ax1.plot(x, y)
ax1.set_title("Graph 1")

ax2.plot(x, y)
ax2.set_title("Graph 2")

ax3.plot(x, y)
ax3.set_title("Graph 3")

ax4.plot(x, y)
ax4.set_title("Graph 4")

plt.tight_layout()

plt.show()
```
